```python
import pandas as pd
import numpy as np
```

```python
books = pd.read_csv("data/books.csv")
train = pd.read_csv("data/train_ratings.csv")
test = pd.read_csv("data/test_ratings.csv")
users = pd.read_csv("data/users.csv")
sub = pd.read_csv("data/sample_submission.csv")
```

```python
train_isbn= train['isbn'].value_counts().to_frame()
```

```python
test_isbn= test['isbn'].value_counts().to_frame()
```

```python
test_isbn.reset_index(inplace=True)
```

```python
train_isbn.reset_index(inplace=True)
```

```python
pd.concat([train_isbn, test_isbn], axis=1)
```

```python
df = train_isbn.merge(books, how='left', left_on = 'index', right_on = 'isbn')
train_book_title = df['book_title'].value_counts().to_frame().reset_index()
```

```python
df1 = test_isbn.merge(books, how='left', left_on = 'index', right_on = 'isbn')
test_book_title = df1['book_title'].value_counts().to_frame().reset_index()
```
