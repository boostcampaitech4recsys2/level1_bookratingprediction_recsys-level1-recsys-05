```python
import pandas as pd
import numpy as np
from scipy.sparse import csr_matrix, linalg
# !pip install xgboost
from xgboost import XGBRegressor, XGBClassifier
# !pip install lightgbm
from lightgbm import LGBMRegressor, LGBMClassifier
# !pip install catboost
from catboost import CatBoostRegressor, CatBoostClassifier, Pool
from sklearn.metrics import accuracy_score
from sklearn.metrics import mean_squared_error
import warnings
from sklearn.model_selection import train_test_split
warnings.filterwarnings(action='ignore')
```

```python
newtrain = pd.read_csv("new_train.csv")
```

```python
seed=42
```

```python
newtrain.head()
```

```python
newtrain['rating'] = newtrain['rating'] -1
```

```python
X_train_cat, X_test_cat, y_train_cat, y_test_cat = train_test_split(newtrain.drop(['user_id', 'isbn', 'rating'], axis=1), newtrain['rating'], test_size=0.2, shuffle=True, random_state=seed)
```

## second EDA

```python
catboost_cl = CatBoostClassifier()
```

```python
catboost_cl.fit(X_train_cat, y_train_cat)
```

```python
# Predict
preds = catboost_cl.predict(X_test_cat)
```

```python
preds
```

```python
accuracy_score(y_test_cat, preds)
```

```python
cbr= CatBoostRegressor()
```

```python
cbr.fit(X_train_cat, y_train_cat)
```

```python
cbr
```

```python
preds = cbr.predict(X_test_cat)
mean_squared_error(y_test_cat, preds)**0.5
```
